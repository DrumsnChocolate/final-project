```python
import torch
import matplotlib.pyplot as plt
from mmengine.model.utils import revert_sync_batchnorm
from mmseg.apis import init_model, inference_model, show_result_pyplot
from mmseg.datasets.ade import ADE20KDataset
```

```python
# config_file = '../configs/setr/setr_vit-l_pup_8xb2-160k_ade20k-512x512.py'
config_file = '../configs/setr/setr_vit-l_pup_8xb2-160k_ade20k-512x512.py'
checkpoint_file = '../checkpoints/setr_pup_512x512_160k_b16_ade20k_20210619_191343-7e0ce826.pth'
```

```python
# build the model from a config file and a checkpoint file
model = init_model(config_file, checkpoint_file, device='cuda')
```

```python
# test a single image
img = 'demo.png'
if not torch.cuda.is_available():
    model = revert_sync_batchnorm(model)
result = inference_model(model, img)
```

```python
# show the results
vis_result = show_result_pyplot(model, img, result, show=False)
plt.imshow(vis_result)
```
