```python
import glob
import os
```

```python

def get_patience(file):
    for line in file.readlines():
        if "'PATIENCE':" in line:
            patience = line.split("'PATIENCE':")[1].split(",")[0].strip()
            return int(patience)
    return None

folder_paths = glob.glob("*/sup_vitb16_imagenet21k/*/*/*/*/*/*")
for folder_path in folder_paths:
    filenames_in_folder = os.listdir(folder_path)
    if not 'logs.txt' in filenames_in_folder:
        # print(f'no logs found in {folder_path}')
        continue
    patience = None
    with open(os.path.join(folder_path, 'logs.txt'), 'r') as f:
        patience = get_patience(f)
    new_folder_path = os.path.join(folder_path.split("/run")[0], f"patience{patience}", "run" + folder_path.split("/run")[1])
    os.makedirs(new_folder_path, exist_ok=True)
    os.rename(folder_path, new_folder_path)
    print(folder_path)
    print(new_folder_path)
```
